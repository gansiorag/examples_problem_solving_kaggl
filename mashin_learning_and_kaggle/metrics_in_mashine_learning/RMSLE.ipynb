# RMSLE - Root Mean Squared Logarithmic Error
Среднеквадратичная логарифмическая ошибка


```python
import random
# Генеартор массивов
# генерируем список действительных случайных чисел kol - количество чисел, order -порядок чисел
def list_float(kol,order):
    rezult = [round(random.random()*(10**order),2) for dd in range(kol)]
    return rezult
```

```python
# Массив фактических  значений модели
X = list_float(100,3)
print(X)
```

```python
# Массив предсказанных значений модели
Y = list_float(100,3)
print(Y)
```

# Разные способы расчета RMSLE

```python
from sklearn.metrics import mean_squared_log_error
```

```python
print(mean_squared_log_error(X,Y))
```

```python
# Можно получить оценку нескольких моделей
X = [[0.5, 1], [1, 2], [7, 6]]
Y = [[0.5, 2], [1, 2.5], [8, 8]]
print(mean_squared_log_error(X,Y, multioutput='raw_values'))
```
